```python
# uploading the model
from tensorflow.keras.models import load_model
import cv2

def prediction(image):
  model = load_model('/content/FOOD101_model.h5')
  food = ['burger', 'butter_naan', 'chai', 'chapati', 'chole_bhature', 'dal_makhani', 'dhokla', 'fried_rice', 'idli', 'jalebi', 'kathi_roll', 'kadhai_paneer', 'kulfi', 'masala_dosa', 'momos', 'paani_puri', 'pakode', 'pav_bhaji', 'pizza', 'samosa']
  #processing the image
  img_arr = cv2.imread(image)
  img = cv2.resize(img_arr,(64,64))/255.0
  img = np.expand_dims(img, axis=0)
  predictions = model.predict(img)
  confidence_scores = predictions[0]
  predicted_class_index = np.argmax(confidence_scores)
  predicted_class = food[predicted_class_index]
  confidence_percentage = confidence_scores[predicted_class_index] * 100

  print(f'the given image is of {predicted_class} with confidence {confidence_percentage}')
```

```python
prediction('/content/pav_bhaji.png')
image = mpimg.imread('/content/pav_bhaji.png')
plt.imshow(image)
plt.axis('off')
plt.show()
```
